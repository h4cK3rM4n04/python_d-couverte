Voyons ci-dessous un exemple plus complexe, qui permet d'afficher la population de Madagascar par district. 
Vous savez maintenant où se trouve le bouton <code>Exécuter</code>

```python
# fichier obtenu sur http://www.resiliencemada.gov.mg/layers/geonode:population_mada/metadata_detail
# données 2015
import folium
import webbrowser
import csv

# importation au format csv
csvfile = open('population_mada.csv')
population_csv = csv.reader(csvfile, delimiter=',',)

# Création de la carte centrée sur 45°N 1°W
carte_stations = folium.Map(location=[-18.9,47], zoom_start=6)

i = 0
for ligne in population_csv :
    if i!=0 :
        longitude = float(ligne[5])
        latitude = float(ligne[6])
        population = int(ligne[7])
        nom = str(ligne[2])
        nom = nom + " : " + str(population) + " hab"
        if population > 200000 : 
            folium.Marker([latitude,longitude], popup = nom, icon = folium.Icon(color='red')).add_to(carte_stations)
        elif population > 100000 : 
            folium.Marker([latitude,longitude], popup = nom, icon = folium.Icon(color='blue')).add_to(carte_stations)
        elif population > 50000 : 
            folium.Marker([latitude,longitude], popup = nom, icon = folium.Icon(color='green')).add_to(carte_stations)
        elif population > 30000 : 
            folium.Marker([latitude,longitude], popup = nom, icon = folium.Icon(color='orange')).add_to(carte_stations)
    else :
        i+=1

carte_stations
```
